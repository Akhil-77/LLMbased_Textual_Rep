```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np

def row_to_text(row):
    return (
        f"{row['Gender']} customer aged {int(row['Age'])} from {row['Geography']} "
        f"with a credit score of {row['CreditScore']}, tenure of {row['Tenure']} years, "
        f"balance {row['Balance']}, has {row['NumOfProducts']} product(s), "
        f"{'has' if row['HasCrCard'] == 1 else 'does not have'} a credit card, "
        f"{'is' if row['IsActiveMember'] == 1 else 'is not'} an active member, "
        f"and has an estimated salary of {row['EstimatedSalary']}."
    )

def generate_embeddings(input_csv_path: str, output_embedding_path: str = None):
    df = pd.read_csv(input_csv_path)
    df = df.drop(columns=['id', 'CustomerId', 'Surname'], errors='ignore')
    df['text_description'] = df.apply(row_to_text, axis=1)
    model = SentenceTransformer('all-MiniLM-L6-v2')

    print("Generating embeddings...")
    embeddings = model.encode(df['text_description'].tolist(), show_progress_bar=True)

    # Optionally save embeddings
    if output_embedding_path:
        np.save(output_embedding_path, embeddings)
        print(f"Embeddings saved to {output_embedding_path}")

    return embeddings, df['text_description']

if __name__ == "__main__":
    embeddings, descriptions = generate_embeddings("/content/train.csv", "embeddings.npy")
```
